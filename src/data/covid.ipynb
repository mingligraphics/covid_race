```python
import pandas as pd
```

```python
df = pd.read_csv("Weekly_Updates.csv")
```

```python
df = df[['State','Indicator','Non-Hispanic Black or African American']]
```

```python
df = df[3:]
```

```python
df
```

```python
df_c = df[df['Indicator'] == 'Distribution of COVID deaths (%)']
```

```python
df_p = df[df.Indicator == 'Unweighted distribution of population (%)']
```

```python
df_c = df_c[['State','Non-Hispanic Black or African American']]
```

```python
df_p = df_p[['State','Non-Hispanic Black or African American']]
```

```python
df_c.columns = ['State','covid']
```

```python
df_p.columns = ['State','pop']
```

```python
merge = pd.merge(df_c,df_p,on='State')
```

```python
merge
```

```python
merge.to_csv("state_covid",index=False)
```
